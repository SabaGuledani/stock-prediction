```python
import os
import pandas as pd
```

```python
filenames = os.listdir('stocks/')
```

```python
files_to_delete = []
for file in filenames:
    df = pd.read_csv("stocks/"+file)
    if len(df) < 3300:
        files_to_delete.append(file)
```

```python
for file in files_to_delete:
    os.remove("stocks/" + file)
```

```python

for file in filenames:
    df = pd.read_csv("stocks/"+file)
    df = df[len(df) - 3300 :]
    df.to_csv("stocks/"+file)
    
```
